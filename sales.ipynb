```python
#1a.Load the data into a Pandas DataFrame and display the first five rows.
s_d=pd.read_csv("C:/Users/eshwa/Downloads/Assignment_2/sales_data_sample.csv", encoding='ISO-8859-1')
s_d.head(5)
```

```python
#1b.Display the column names and their data types.
s_d.dtypes
```

```python
#1c.Count the total number of unique ORDERNUMBER values.
s_d['ORDERNUMBER'].drop_duplicates().count()
```

```python
#2a.Filter and display only the orders where the STATUS is "Shipped" and PRODUCTLINE is "Motorcycles."
filtered_orders = s_d[(s_d['STATUS'] == 'Shipped') & (s_d['PRODUCTLINE'] == 'Motorcycles')]
filtered_orders
```

```python
#2b.Find all orders with a SALES value greater than 4000 and display their ORDERNUMBER,SALES, and ORDERDATE
order_sales = s_d[s_d['SALES'] > 4000][['ORDERNUMBER','SALES','ORDERDATE']]
order_sales
```

```python
#2c.Display orders from the year 2004 with a QUANTITYORDERED greater than 30.
orderss = s_d[(s_d['YEAR_ID'] == 2004 ) & (s_d['QUANTITYORDERED'] > 30)]
orderss
```

```python
#3a.Calculate the total SALES for each YEAR_ID and PRODUCTLINE.
total_sales=s_d.groupby(['YEAR_ID','PRODUCTLINE'])['SALES'].sum()
total_sales

```

```python
#3b.Find the average PRICEEACH for each unique PRODUCTLINE.
PRICEEACH = s_d.groupby(['PRODUCTLINE'])['SALES'].mean()
PRICEEACH
```

```python
#3c.Identify the month (MONTH_ID) with the highest SALES in 2003.
sales_2003= s_d[s_d['YEAR_ID']== 2003]
highest_sale = sales_2003.sort_values(by='SALES',ascending= False).head(1)
highest_sale['MONTH_ID']
```

```python
#4a.Sort the DataFrame by ORDERDATE in ascending order and display the first five rows.
s_d['ORDERDATE'].sort_values().head(5)
```

```python
#4b.Find the top 3 orders with the highest SALES and display their ORDERNUMBER, SALES, and PRODUCTLINE.
highest_sales=s_d.sort_values(by='SALES', ascending=True).head(3)[['ORDERNUMBER', 'SALES', 'PRODUCTLINE']]
highest_sales
```

```python
#5a Create a new column called TOTAL_PRICE that calculates the total price for each order(QUANTITYORDERED * PRICEEACH)
s_d['TOTAL_PRICE']= s_d['QUANTITYORDERED']*s_d['PRICEEACH']
s_d['TOTAL_PRICE']
```

```python
#5b Convert the ORDERDATE column to a datetime format and extract the month name, saving it in a new column called MONTH_NAME.
convert=pd.to_datetime(s_d['ORDERDATE'])
s_d['MONTH_NAME'] = convert.dt.month_name()
s_d['MONTH_NAME']
```

```python
#5c.Replace any PRODUCTLINE labeled "Classic Cars" with "Vintage Cars"
replace = s_d['PRODUCTLINE'].replace('Classic Cars','Vintage Cars')
replace
```

```python
#6a.Group the data by YEAR_ID and calculate the total SALES for each year
total_sales_by_year = s_d.groupby(['YEAR_ID'])['SALES'].sum()
total_sales_by_year
```

```python
#6b.Use a pivot table to display the total SALES for each PRODUCTLINE by QTR_ID
pivot = pd.pivot_table(s_d,
                      index ='QTR_ID',
                      columns='PRODUCTLINE',
                      values='SALES',
                      aggfunc='sum'
                      
)
pivot
```

```python
#6c. Find the MSRP average for each PRODUCTLINE across all orders.
msrp_average = s_d.groupby(['PRODUCTLINE'])['MSRP'].mean()
msrp_average
```

```python
#7a. What is the total quantity ordered for each PRODUCTLINE?
total_quantity = s_d.groupby(['PRODUCTLINE'])['QUANTITYORDERED'].sum()
total_quantity
```

```python
#7b. How many orders were placed each quarter (QTR_ID) in 2003?
orders_by_each_quater = s_d[s_d['YEAR_ID']==2003]['QTR_ID'].sum()
orders_by_each_quater
```

```python
#7c. Which ORDERNUMBER had the highest SALES in 2005?
highest_sale_order_num = s_d[s_d['YEAR_ID'] == 2005].sort_values(by ='SALES',ascending = False).head(1)['ORDERNUMBER']
highest_sale_order_num
```
